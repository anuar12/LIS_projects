```python
from __future__ import print_function
import numpy as np
np.random.seed(1337)  # for reproducibility
from sklearn import decomposition
import matplotlib.pyplot as plt
from ggplot import *
import pandas as pd

from keras.datasets import mnist
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.optimizers import SGD, Adam, RMSprop
from keras.utils import np_utils
from keras.layers.convolutional import Convolution2D, MaxPooling2D, Convolution1D
```

```python
import pandas as pd
train = pd.read_hdf("/Users/Anuar_The_Great/desktop/project/4/train_labeled.h5", "train")
train_un = pd.read_hdf("/Users/Anuar_The_Great/desktop/project/4/train_unlabeled.h5", "train")
test = pd.read_hdf("/Users/Anuar_The_Great/desktop/project/4/test.h5", "test")
```

```python
print(train.shape)
print(train_un.shape)
print(test.shape)
test
```

```python
X_train = np.array(train.values)[0:7000, 1:130]
Y_train = np.array(train.values)[0:7000, 0].reshape(7000, 1)
X_test = np.array(train.values)[7000:, 1:130]
Y_test = np.array(train.values)[7000:, 0].reshape(2000, 1)
X_test_submit = np.array(test.values)

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
print(X_train.shape)
print(Y_test.shape)
print(X_test_submit.shape)
```

```python
#Normalizing the training data, transforming the X_test and actual test data accordingly
#training samples 40k
X_train -= np.mean(X_train, axis = 0) # zero-center the data (important)
#X_train /= np.std(X_train, axis = 0)  # normalization: can't do because of zero std for some columns

#validation samples 5324
X_test -= np.mean(X_test, axis = 0) # zero-center the data (important)
#X_test /= np.std(X_train, axis = 0)  # normalization: can't do because of zero std for some columns

#actual test samples
X_test_submit -= np.mean(X_test_submit, axis = 0) # zero-center the data (important)
#test /= np.std(X_train, axis = 0)  # normalization: can't do because of zero std for some columns
```

```python
pca = decomposition.PCA()
pca.n_components = 80
Xtrain_reduced = pca.fit_transform(X_train)  #This is the one we're going to train on
Xtest_reduced = pca.transform(X_test)  #This is the one from the train dataset which we'll use for testing
Xtest_submit_reduced = pca.transform(X_test_submit)
testdata_forprediction = pca.transform(np.array(test.values))  #This is the transformed test data which we'll use for prediction
print('Size of Xtrain_reduced = ', Xtrain_reduced.shape)
print('Size of Xtest_reduced = ', Xtest_reduced.shape)
print('Size of testdata_forprediction = ', testdata_forprediction.shape)
```

```python
pca.explained_variance_ratio_*100
```

```python
nb_classes = 10
batch_size = 128


Y_train = np_utils.to_categorical(Y_train, nb_classes)
Y_test = np_utils.to_categorical(Y_test, nb_classes)
```

```python
model = Sequential()
model.add(Dense(456, input_shape=(80,), init='lecun_uniform'))
model.add(Activation('relu'))
model.add(Dropout(0.3))
model.add(Dense(456, init='lecun_uniform'))
model.add(Activation('relu'))
model.add(Dropout(0.3))
model.add(Dense(10, init='lecun_uniform'))
model.add(Activation('softmax'))

model.summary()
```

```python
model2 = Sequential()
model2.add(Dense(300, input_shape=(80,), init='lecun_uniform'))
model2.add(Activation('relu'))
model2.add(Dropout(0.3))
model2.add(Dense(150, init='lecun_uniform'))
model2.add(Activation('relu'))
model2.add(Dropout(0.3))
model2.add(Dense(10, init='lecun_uniform'))
model2.add(Activation('softmax'))

model2.summary()
```

```python
model3 = Sequential()
model3.add(Dense(500, input_shape=(80,), init='lecun_uniform'))
model3.add(Activation('relu'))
model3.add(Dropout(0.4))
model3.add(Dense(400, init='lecun_uniform'))
model3.add(Activation('relu'))
model3.add(Dropout(0.4))
model3.add(Dense(10, init='lecun_uniform'))
model3.add(Activation('softmax'))

model3.summary()
```

```python
nb_epoch = 20

model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history = model.fit(Xtrain_reduced, Y_train,
                    batch_size=batch_size, nb_epoch=nb_epoch,
                    verbose=2, validation_data=(Xtest_reduced, Y_test), shuffle=True)
score = model.evaluate(Xtest_reduced, Y_test, verbose=0)
print('Test score:', score[0])
print('Test accuracy:', score[1])
```

```python
nb_epoch = 15

model2.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history = model2.fit(Xtrain_reduced, Y_train,
                    batch_size=batch_size, nb_epoch=nb_epoch,
                    verbose=2, validation_data=(Xtest_reduced, Y_test), shuffle=True)
score = model2.evaluate(Xtest_reduced, Y_test, verbose=0)
print('Test score:', score[0])
print('Test accuracy:', score[1])
```

```python
nb_epoch = 15

model3.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])

history = model3.fit(Xtrain_reduced, Y_train,
                    batch_size=batch_size, nb_epoch=nb_epoch,
                    verbose=2, validation_data=(Xtest_reduced, Y_test), shuffle=True)
score = model3.evaluate(Xtest_reduced, Y_test, verbose=0)
print('Test score:', score[0])
print('Test accuracy:', score[1])
```
